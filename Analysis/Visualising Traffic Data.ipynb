# Visualising one at a time

```python
uniqueUsers=df.site.unique()
DataFrameDict = {elem : pd.DataFrame for elem in uniqueUsers}
for key in DataFrameDict.keys():
        DataFrameDict[key]=df[:][df.site==key]
```

```python
df_tesco=DataFrameDict['https://www.tesco.com/']

ax=plt.gca()

ax.plot(df_tesco.date, df_tesco['Alexa Rank'] )
plt.xlabel('Date')
plt.ylabel('Alexa Web Rank')

plt.title('Tesco')
```
