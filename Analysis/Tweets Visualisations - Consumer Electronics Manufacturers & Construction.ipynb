```python
import pandas as pd
import glob
```

```python
files=glob.glob('../Data/Tweets/*.csv')

df=(pd.read_csv(f) for f in files)
totaldf=pd.concat(df, ignore_index=True)
```

```python
totaldf.head()
```

```python
totaldf['Date'], totaldf['Time']=totaldf['created_at'].str.split(' ', 1).str
totaldf=totaldf.drop({'created_at'},1)
```

```python
totaldf.head()
```

# Freq by date

```python
import datetime

import matplotlib.pyplot as plt
% matplotlib inline

from matplotlib.dates import YearLocator, MonthLocator, DateFormatter
```

```python
years = YearLocator()   # every year
months = MonthLocator()  # every month
yearsFmt = DateFormatter('%Y')
```

```python
totaldf['Date']=pd.to_datetime(totaldf['Date'])
```

```python
uniqueUsers=totaldf.User.unique()
DataFrameDict = {elem : pd.DataFrame for elem in uniqueUsers}
for key in DataFrameDict.keys():
        DataFrameDict[key]=totaldf[:][totaldf.User==key]
    
```

# Consumer electronics manufacturers

```python
df_GP=DataFrameDict['GoPro']
df_GP.index=df_GP['Date']

df_GPcount=df_GP.resample('W').count()

ax=plt.gca()
ax.xaxis.set_major_locator(years)
ax.xaxis.set_major_formatter(yearsFmt)
ax.xaxis.set_minor_locator(months)
ax.autoscale_view()

plt.plot(df_GPcount.index, df_GPcount.text, 'k')
plt.ylabel('Number of tweets')
plt.title('@GoPro tweets per week')
```
